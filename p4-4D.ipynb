```python
# define variables
alpha = cp.Variable(m)
C = 1.0

# Define objective and constraints
objective = cp.Maximize(cp.sum(alpha) - 
                        0.5 * cp.quad_form(alpha * np.outer(b, b), 
                                           np.dot(diabetes_X, diabetes_X.T)))
constraints = [cp.sum(alpha * b) == 0, alpha >= 0, alpha <= C]

problem = cp.Problem(objective, constraints)
problem.solve()

optimal_alpha = alpha.value

# Calculate the weights
w = np.sum(optimal_alpha.reshape(-1, 1) * b * diabetes_X, axis=1) #442x1 442x1 442x10
print(w)
```
